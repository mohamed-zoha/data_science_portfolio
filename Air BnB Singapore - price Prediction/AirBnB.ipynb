```python
dct = {}
for i in df.amenities:
    j = i[1:-1:].split(",")
    for k in j:
        if k in dct.keys():
            dct[k] +=1
        else:
            dct[k] = 1
```

```python
amenities = pd.DataFrame(dct.values(),index=dct.keys()).sort_values(0,ascending=False).reset_index().rename({"index":"amenities",
                                                                                                0:"frequencies"},axis=1)
```
